```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#load data
df = pd.read_csv('/content/ball_by_ball_it20.csv')
```

```python
df
```

```python
#data preprocessing
# Check for missing values
missing_values = df.isnull().sum()

# Convert date columns to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove duplicate rows
df = df.drop_duplicates()

# Check data types
df.info()
```

```python
# Summary statistics
print(df.describe())
```

```python
print(df.columns)
```

```python
plt.hist(df['Batter Runs'], bins=20)
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.title('Distribution of Runs Scored')
plt.show()

```

```python
sns.boxplot(x='Innings', y='Batter Runs', data=df)
plt.xlabel('Innings')
plt.ylabel('Runs Scored')
plt.title('Runs Scored by Innings')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a correlation heatmap with improved aesthetics
plt.figure(figsize=(12, 10))
sns.set(font_scale=0.6)

# Customize the color palette for better visualization
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Create the heatmap with annotations and color mapping
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap=cmap, center=0)

plt.title('Correlation Heatmap', fontsize=26)
plt.show()


```

```python
plt.scatter(df['Batter Runs'], df['Innings Wickets'])
plt.xlabel('Runs Scored')
plt.ylabel('Wickets Taken')
plt.title('Scatterplot: Runs vs. Wickets')
plt.show()
```

```python
#bar chart

winner_counts = df['Winner'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(winner_counts.index, winner_counts)
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Wins by Team')
plt.xticks(rotation=90)
plt.show()

```

```python
#time series plot
df['Date'] = pd.to_datetime(df['Date'])

time_series_df = df.groupby('Date')['Innings Runs'].sum()

plt.figure(figsize=(12, 6))
plt.plot(time_series_df.index, time_series_df.values, marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Total Innings Runs')
plt.title('Time Series of Innings Runs')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
